```python
#import dependencies 
import pandas as pd
import time
import os
from bs4 import BeautifulSoup as bs
from splinter import Browser
from selenium import webdriver
```

```python
chrome_path= '/Users/vaishnaviiparamashivam/Desktop/chrom/chromedriver'
```

```python
browser = webdriver.Chrome(chrome_path)
```

```python
# initialize path /browser
executable_path = {"executable_path": "chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# URL of page to be scraped 
url = "https://www.imsdb.com/all%20scripts/"
browser.visit(url)
time.sleep(1)
```

```python
# beautiful soup object
html = browser.html
soup = bs(html, 'html.parser')
```

```python
#test to see if the code works- scraping the first movie title 
titles = soup.find('p', title_='').text
print(titles)
```

```python
#list of movies in IMSDB
movies = soup.find_all('a', class_='')

#create list
movie_title = []
movie_links= []

for movie in movies:
    title = movie.text
    movie_title.append(title)
```

```python
#printing list to check 
movie_title
```

```python
#counting how many values are in the list 
len(movies)
```

```python
# scraping URL list 
movieurls = soup.find_all('a', class_='')
for x in movieurls:
    links = x.get('href')
    movie_links.append(links)
```

```python
movie_links
```

```python
#checking to see if the length matches the movie titles 
len(movie_links)
```

```python
#list to dataframe. setting index for merge
title_df = pd.DataFrame(movie_title)
title_df.reset_index(level=0, inplace=True)
title_df.head()
```

```python
#list to dataframe. setting index for merge
link_df = pd.DataFrame(movie_links)
link_df.reset_index(level=0, inplace=True)
link_df.head()
```

```python
#merging dataframes , renaming the columns 
movies_df = pd.merge(title_df, link_df, on='index')
movies_df = movies_df.rename(columns ={'0_x':"Movie_Title", '0_y': "Movie_Links"})
movies_df.head()
```

```python
#output datafram to csv file for easier clean up 
movies_df.to_csv("/Users/vaishnaviiparamashivam/Documents/Badwords_corr_proj/output/movie_list.csv")
```
